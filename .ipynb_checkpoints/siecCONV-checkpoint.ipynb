```python
# Importowanie bibliotek
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sklearn.metrics as skm
import sklearn.model_selection as skms
import sklearn.preprocessing as skp
import librosa
import librosa.display as lplt
import tensorflow as tf
import keras as k
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import IPython

# Ustawienia
sns.set_style('darkgrid')
%matplotlib inline
warnings.filterwarnings('ignore')
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)

# Wczytywanie danych
data_frame = pd.read_csv('C:/Users/jangr/Documents/sieci2_data/features_3_sec.csv')
print("Dataset shape:", data_frame.shape)
print("Count of each label:")
print(data_frame.label.value_counts().reset_index())

# Wizualizacja danych
audio_file_path = 'C:/Users/jangr/Documents/sieci2_data/genres_original/pop/pop.00005.wav'
audio_data, sample_rate = librosa.load(audio_file_path)
audio_data, _ = librosa.effects.trim(audio_data)

# Odtworzenie pliku audio
IPython.display.Audio(audio_data, rate=sample_rate)


# Normalizacja danych
data_features = data_frame.iloc[:, 1:]
y_labels = data_features['label']
X_features = data_features.loc[:, data_features.columns != 'label']
min_max_scaler = MinMaxScaler()
X_features_scaled = pd.DataFrame(min_max_scaler.fit_transform(X_features), columns=X_features.columns)


# Przygotowanie danych
print("Columns with NA values:", list(data_frame.columns[data_frame.isnull().any()]))
label_index_map = {label: idx for idx, label in enumerate(data_frame.label.unique())}
index_label_map = {idx: label for label, idx in label_index_map.items()}
print(label_index_map)
print(index_label_map)
data_frame.label = data_frame.label.map(label_index_map)

# Losowanie próbek
data_frame_shuffled = data_frame.sample(frac=1, random_state=seed).reset_index(drop=True)
data_frame_shuffled.drop(['filename', 'length'], axis=1, inplace=True)
data_frame_y = data_frame_shuffled.pop('label')
data_frame_X = data_frame_shuffled

# Podział danych na zbiory
X_train, df_test_valid_X, y_train, df_test_valid_y = skms.train_test_split(data_frame_X, data_frame_y, train_size=0.7, random_state=seed, stratify=data_frame_y)
X_dev, X_test, y_dev, y_test = skms.train_test_split(df_test_valid_X, df_test_valid_y, train_size=0.66, random_state=seed, stratify=df_test_valid_y)

print(f"Zbiór treningowy: {X_train.shape[0]} z {len(data_frame_shuffled)}")
print(f"Zbiór walidacyjny: {X_dev.shape[0]} z {len(data_frame_shuffled)}")
print(f"Zbiór testowy: {X_test.shape[0]} z {len(data_frame_shuffled)}")

# Normalizacja
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_dev = pd.DataFrame(scaler.transform(X_dev), columns=X_dev.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

# Definiowanie i trenowanie modelu CNN
class CustomCallback(k.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        if logs.get('val_accuracy') > ACCURACY_THRESHOLD:
            print(f"\n\nZatrzymanie trenowania po osiągnięciu {ACCURACY_THRESHOLD * 100:.2f}% dokładności!")
            self.model.stop_training = True

def train_cnn_model(model, epochs, optimizer):
    batch_size = 256
    callback = CustomCallback()
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model.fit(X_train, y_train, validation_data=(X_dev, y_dev), epochs=epochs, batch_size=batch_size, callbacks=[callback])

def plot_training_history(history):
    print("Maks. dokładność walidacyjna:", max(history.history["val_accuracy"]))
    pd.DataFrame(history.history).plot(figsize=(12, 6))
    plt.show()

# Budowanie modelu CNN
cnn_model = k.models.Sequential([
    k.layers.Conv1D(64, 2, activation='relu', input_shape=(X_train.shape[1], 1)),
    k.layers.MaxPooling1D(2),
    k.layers.Conv1D(128, 2, activation='relu'),
    k.layers.MaxPooling1D(2),
    k.layers.Conv1D(256, 3, activation='relu'),
    k.layers.GlobalAveragePooling1D(),
    k.layers.Dense(512, activation='relu', kernel_regularizer=k.regularizers.l2(0.001)),
    k.layers.Dropout(0.5),
    k.layers.Dense(256, activation='relu', kernel_regularizer=k.regularizers.l2(0.001)),
    k.layers.Dropout(0.5),
    k.layers.Dense(128, activation='relu', kernel_regularizer=k.regularizers.l2(0.001)),
    k.layers.Dropout(0.5),
    k.layers.Dense(64, activation='relu', kernel_regularizer=k.regularizers.l2(0.001)),
    k.layers.Dense(10, activation='softmax')
])

# Wyświetlenie podsumowania modelu
print(cnn_model.summary())

# Trenowanie modelu
cnn_model_history = train_cnn_model(cnn_model, epochs=200, optimizer='adam')

# Wyświetlenie historii trenowania
plot_training_history(cnn_model_history)

# Ocena modelu
test_loss, test_acc = cnn_model.evaluate(X_test, y_test, batch_size=256)
print("Straty:", test_loss)
print("Najlepsza dokładność:", test_acc * 100)

# Predykcje
predictions = cnn_model.predict(X_test)

# Wyświetlenie przewidywań
index_example = 50
example_percentages = predictions[index_example]
predicted_genre_index = np.argmax(example_percentages)
predicted_genre = index_label_map[predicted_genre_index]
print(f"Przewidywany gatunek dla indeksu {index_example}: {predicted_genre}")

print("Wartości procentowe dla wybranego przykładu:")
for i, percentage in enumerate(example_percentages):
    genre = index_label_map[i]
    print(f"{genre}: {percentage * 100:.2f}%")

print("\nPrzypisane gatunki dla wartości 0-9:")
for i in range(10):
    genre_index = np.argsort(example_percentages)[-i - 1]
    genre = index_label_map[genre_index]
    print(f"{i}: {genre}")
```
